```python
# rita grafer 

'''
    Ett flygplan håller en konstant hastighet av 800km/h. Efter x h har det hunnit y km
    
    Värdetabell
    x, y
    0, 0
    1, 800
    2, 1600
    3, 2400
    4, 3200
    5, 4000
'''
import matplotlib.pyplot as plt 

# lösning 1 med listor
x = [0,1,2,3,4,5]
y = [0,800,1600,2400,3200,4000]

plt.plot(x,y, 'o-')
plt.title("Flygplan med konstanta hastigheten 800km/h")
plt.xlabel("Tid i timmar")
plt.ylabel("Sträckan i km")
plt.show()
```

```python
import math 
import random as rnd
import matplotlib.pyplot as plt


inne,ute = 0,0

for i in range(2000):
    x = rnd.uniform(-1,1)
    y = rnd.uniform(-1,1)

    d = math.sqrt(x**2+y**2)

    if d <= 1:
        inne += 1
        plt.plot(x,y,".g")
    else: 
        ute += 1
        plt.plot(x,y,".r")

inne/i*4
plt.gca().set_aspect('equal')
```
